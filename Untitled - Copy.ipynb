```python
import pandas as pd
import numpy as np
import gzip
import json
import os
import sys
```

```python
def load_data(file_name, head = 500):
    count = 0
    data = []
    with gzip.open(file_name) as fin:
        for l in fin:
            d = json.loads(l)
            count += 1
            data.append(d)
            
            # break if reaches the 100th line
            if (head is not None) and (count > head):
                break
    return data
```

```python
load_data('./data/goodreads_interactions_poetry.json.gz')
```
